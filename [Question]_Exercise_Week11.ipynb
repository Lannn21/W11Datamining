## Preprocess Data

In this step, we will transform our dataset so that we will have a one hot encoding based on the purchased products.

```python
unique = pd.unique(df.values.ravel())
encoding = {product: 1 if product not in ['Milk', 'Bagel', np.nan] else 0 for product in unique}

print(encoding)
```

```python
unique_products = pd.unique(df.values.ravel())

encoded_df = pd.DataFrame(0, index=df.index, columns=unique_products)

for index, row in df.iterrows():
    for product in row:
        if product in encoded_df.columns:
            encoded_df.at[index, product] = 1

encoded_df.head()
```

```python
encoded_df = encoded_df.drop(columns=[np.nan])

encoded_df.head()
```

Since, the encoded dataframe consist of the empty column. We will drop the NaN column or select all columns other than the first column.

## Apriori Algorithm

We will use appriori algorithm to determine the frequently purchased products.
For this case study, we will min_support=0.2

```python
from mlxtend.frequent_patterns import apriori, association_rules

frequent_itemsets = apriori(encoded_df, min_support=0.2, use_colnames=True)
frequent_itemsets
```

The we will generate association rule of the frequent itemset based on confidence level with the threshold=0.6

```python
association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6)
```

Antecedent support mengukur seberapa sering item atau kombinasi item di sisi kiri aturan muncul dalam dataset, misalnya, "Bagel" muncul sekitar 42.5% dalam dataset. Consequent support mengukur seberapa sering item di sisi kanan aturan muncul, seperti "Bread" yang muncul 50.5% dalam transaksi. Support mengukur seberapa sering pasangan antecedent dan consequent muncul bersamaan, contohnya, "Bagel  Bread" muncul 27.9% dari transaksi.

Confidence menunjukkan seberapa besar kemungkinan consequent muncul jika antecedent ada, contohnya, jika "Bagel" ada, maka ada 65.7% kemungkinan "Bread" juga ada. Lift membandingkan probabilitas munculnya kedua item bersama-sama dengan probabilitas acak mereka muncul, dan nilai lebih besar dari 1 (seperti 1.3) menunjukkan keduanya lebih sering muncul bersama daripada yang diharapkan secara acak. Leverage mengukur peningkatan yang terjadi pada support bersama dibandingkan dengan ekspektasi jika keduanya muncul secara independen, yang pada contoh ini menunjukkan peningkatan 6.5%.

Conviction mengukur seberapa kuat hubungan antara antecedent dan consequent, dan nilai lebih besar dari 1 (seperti 1.44) menunjukkan hubungan yang kuat. Zhangs metric adalah metrik lain yang mengevaluasi seberapa relevan aturan asosiasi tersebut, dengan nilai 0.40 menunjukkan aturan tersebut memiliki kekuatan yang cukup baik, meskipun tidak terlalu kuat.

Secara keseluruhan, aturan asosiasi yang ditemukan menunjukkan hubungan yang signifikan antara produk-produk tertentu, seperti "Bagel" dan "Bread", yang muncul lebih sering bersama-sama daripada yang diharapkan secara acak.
